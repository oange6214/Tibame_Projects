```
%tensorflow_version 2.x

from tensorflow.keras.preprocessing import image
import tensorflow as tf

from glob import glob
import os, zipfile
import pandas as pd
import numpy as np

# tf.keras.backend.set_learning_phase(True)
if not os.path.exists('C1-P1_Test'):
    f = zipfile.ZipFile("/content/drive/My Drive/Colab Notebooks/C1-P1_Test.zip")
    f.extractall("./")
```

```
FOLDERS = '/content/drive/My Drive/'
DATASET_PATH = '/content/C1-P1_Test'
IMAGE_SIZE = 224
NUM_CLASSES = 3
```

```
import efficientnet.tfkeras as efn
```

```
imgPaths = [path for path in sorted(glob(os.path.join(DATASET_PATH, '*')))]

model = tf.keras.models.load_model(os.path.join(FOLDERS, 'mango_v9-2.h5'))
```

```
imgList = []
# [-1, 1]
def preprocess_fn1(im_array):
    return im_array / 127.5 -1

# [ 0, 1]
def preprocess_fn2(im_array):
    return im_array / 255

# 平均 / 標準差
def preprocess_fn3(im_array):
    return (im_array - np.mean(im_array))/ np.std(im_array)

for imgPath in imgPaths:
    img = image.load_img(imgPath, target_size=(IMAGE_SIZE, IMAGE_SIZE))
    x = image.img_to_array(img)
    x = preprocess_fn1(x)
    imgList.append(x)

imgList = np.array(imgList)
```

```
answer = ['A', 'B', 'C']

y_pred = np.argmax(model.predict(imgList), axis=-1)
ans = np.array([answer[i] for i in y_pred])
```

```
df = pd.read_csv('/content/test_example.csv')
df['label'] = ans
df.to_csv('v9-2.csv', index=False)
```
