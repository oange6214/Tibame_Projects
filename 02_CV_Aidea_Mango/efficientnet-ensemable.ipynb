```
imgPaths = [path for path in sorted(glob(os.path.join(DATASET_PATH, '*')))]
```

```
model1 = tf.keras.models.load_model(os.path.join(FOLDERS, 'mango_v6-o.h5'))
model2 = tf.keras.models.load_model(os.path.join(FOLDERS, 'mango_v6.h5'))
model3 = tf.keras.models.load_model(os.path.join(FOLDERS, 'mange_v7.h5'))
model4 = tf.keras.models.load_model(os.path.join(FOLDERS, 'mango_v9.h5'))
model5 = tf.keras.models.load_model(os.path.join(FOLDERS, 'mango_v9-2.h5'))
```

# 載入資料

```
# [-1, 1]
def preprocess_fn1(im_array):
    return im_array / 127.5 -1

imgList = []
for imgPath in imgPaths:
    img = image.load_img(imgPath, target_size=(IMAGE_SIZE, IMAGE_SIZE))
    x = image.img_to_array(img)
    x = preprocess_fn1(x)
    imgList.append(x)
imgList = np.array(imgList)
```

```
pre_1 = model1.predict(imgList)
pre_2 = model2.predict(imgList)
pre_3 = model4.predict(imgList)
pre_4 = model4.predict(imgList)
pre_5 = model5.predict(imgList)

pre_n = 0.2 * pre_1 + 0.2 * pre_2 + 0.2 * pre_3 + 0.2 * pre_4 + 0.2 * pre_5
```

```
answer = ['A', 'B', 'C']

y_pred = np.argmax(pre_n, axis=-1)
ans = np.array([answer[i] for i in y_pred])
```

```
df = pd.read_csv('/content/test_example.csv')
df['label'] = ans
df.to_csv('ensemble-1.csv', index=False)
```
